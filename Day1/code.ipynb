```python
sample_input = """
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""
```

```python
def get_rotations(data: str):
    rotations = []
    for line in data.strip().split("\n"):
        if line.startswith("L"):
            rotations.append(-int(line[1:]))
        elif line.startswith("R"):
            rotations.append(int(line[1:]))
        else:
            raise ValueError()
    return rotations
```

```python
def first_part(data):
    x = 0
    position = 50
    for rotation in get_rotations(data):
        position = (position + rotation) % 100

        if position == 0:
            x += 1
    return x
```

```python
import math

def second_part(data):
    x = 0
    position = 50
    for rotation in get_rotations(data):
        new_position = (position + rotation) % 100

        loops = abs(math.floor((position + rotation) / 100))
        if new_position == 0:
            x += 1 + loops
            if rotation > 0:
                x -= 1
        else:
            x += loops
            if position == 0 and rotation < 0:
                x -= 1

        position = new_position 
    return x
```

```python
first_part(sample_input), second_part(sample_input)
```

```python
with open("input.txt", "r") as f:
    puzzle_input = f.read()
print(first_part(puzzle_input), second_part(puzzle_input))
```
