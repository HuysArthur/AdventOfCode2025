```python
sample_input = """
11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124
"""
```

```python
def id_ranges(data: str):
    ranges = []
    for rule in data.strip().split(","):
        ranges.append((rule.split("-")[0], rule.split("-")[1]))
    return ranges
id_ranges(sample_input)
```

```python
from typing import Tuple
def find_invalid_ids_first_part(id_range: Tuple[str, str]):
    id_start, id_end = id_range
    length_diff = len(id_end) - len(id_start)
    id_lengths = [len(id_start) + i for i in range(length_diff + 1)]

    invalid_ids = []
    for id_length in id_lengths:
        if id_length % 2 == 0: # even
            if id_length == len(id_start):
                start = int(id_start[:int(id_length/2)])
            else:
                start = pow(10, int(id_length/2 - 1))

            if id_length == len(id_end):
                end = int(id_end[:int(id_length/2)])
            else:
                end = pow(10, int(id_length/2)) - 1
            
            if int(str(start) * 2) >= int(id_start) and int(str(start) * 2) <= int(id_end): invalid_ids.append(str(start) * 2)
            if start != end and int(str(end) * 2) >= int(id_start) and int(str(end) * 2) <= int(id_end): invalid_ids.append(str(end) * 2)
            invalid_ids.extend([str(i) * 2 for i in range(start + 1, end)])
    
    for invalid_id in invalid_ids:
        try:
            assert int(invalid_id) >= int(id_start) and int(invalid_id) <= int(id_end)
        except:
            print(invalid_id, id_range)
    
    return invalid_ids
```

```python
def first_part(data: str):
    x = 0
    for id_range in id_ranges(data):
        x += sum([int(id) for id in find_invalid_ids_first_part(id_range)])
    return x

with open("input.txt", "r") as f:
    puzzle_input = f.read()

first_part(sample_input), first_part(puzzle_input)
```
